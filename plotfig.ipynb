```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
os.environ['R_HOME'] = 'D:/R-4.4.0'
import scanpy as sc
import scanpy as sc
import anndata as ad
import matplotlib.colors as mcolors
import rpy2.robjects as robjects
from scipy.optimize import minimize
import networkx as nx
from pathlib import Path
from rpy2.robjects import r, pandas2ri
from rpy2.robjects.packages import importr
import pcurve
import heapq
from scipy.spatial.distance import pdist, squareform
from matplotlib.colors import ListedColormap
from matplotlib.colors import LinearSegmentedColormap, to_hex
def amend_matrix(distance_all, neighborsNumber):
    top_ten_indices = np.argsort(distance_all, axis=1)[:, :int(neighborsNumber)]
    distances_transition_proba = np.zeros(distance_all.shape)
    for i in range(len(top_ten_indices)):
        distances_transition_proba[i, top_ten_indices[i]] = 1
    distance_all_amend = np.zeros(distance_all.shape)
    i=0
    for item1 in distances_transition_proba:
        j=0
        for item2 in item1:
            if item2 != 0:
                distance_all_amend[i,j] = distance_all[i,j] 
            j+=1
        i+=1 
    return distance_all_amend, distances_transition_proba
def BuildNet(NodeName, adjacency_matrix):
    # 创建一个无向图
    G = nx.Graph()
    edges = []
    i=0
    for item1 in NodeName:
        j=0
        for item2 in NodeName:
            if adjacency_matrix[i, j] != 0:
                edges.append((item1, item2, {'weight': adjacency_matrix[i, j]}))
            j+=1
        i+=1
    G.add_edges_from(edges)
    return G
def PearsonCorr_func(x, y):
    # 计算均值
    X, Y = x.flatten(), y.flatten()
    mean_X = np.mean(X)
    mean_Y = np.mean(Y)
    # 计算协方差
    cov_XY = np.mean((X - mean_X) * (Y - mean_Y))
    # 计算标准差
    std_X = np.std(X, ddof=0)
    std_Y = np.std(Y, ddof=0)
    # 计算皮尔逊相关系数
    pearson_corr = cov_XY / (std_X * std_Y)
    return pearson_corr
def shortest_path_to_pseudo_time(G, start_node):
    G_copy = G.copy()
    node_list = [start_node]
    for item in list(G_copy.nodes):
        if item not in node_list:
            node_list.append(item)
    node_list_copy = node_list.copy()
    pseudo_time = [0]
    for item in node_list[1:]:
        if nx.has_path(G, start_node, item):
            shortest_path_item = nx.shortest_path(G, source=start_node, target=item, weight='weight')
            pseudo_time_item = []
            for i in range(len(shortest_path_item)-1):
                pseudo_time_item.append(G[shortest_path_item[i]][shortest_path_item[i+1]]['weight'])
            pseudo_time.append(sum(pseudo_time_item))
        else:
            pseudo_time.append(100)
    pseudo_time_df = pd.DataFrame(pseudo_time, columns=['pseudo time1'], index=node_list_copy)
    return pseudo_time_df

def shortest_path_to_pseudo_time_group(G, start_group):
    G_copy = G.copy()
    new_mst_edges = []
    pseudo_time_df = pd.DataFrame(index=list(G_copy.nodes),columns=[str(i) for i in range(len(start_group))])
    i=0
    for item1 in start_group:
        for item2 in list(G_copy.nodes):
            if nx.has_path(G, item1, item2):
                shortest_path_item = nx.shortest_path(G, source=item1, target=item2, weight='weight')
                pseudo_time_df.loc[item2, str(i)] = len(shortest_path_item)
                if len(shortest_path_item) >1:
                    for i in range(len(shortest_path_item)-1):
                        new_mst_edges.append((shortest_path_item[i], shortest_path_item[i+1]))
            else:
                pseudo_time_df.loc[item2, str(i)] = 100
    row_min = pseudo_time_df.min(axis=1)
    row_min_values = row_min.values
    row_min_values = (row_min_values- np.min(row_min_values)) / (np.max(row_min_values)- np.min(row_min_values))
    pseudo_time_df['pseudo time'] = row_min_values
    row_min_df = pseudo_time_df.drop(columns=[str(i) for i in range(len(start_group))])
    new_mst_edges = list(set(new_mst_edges))
    return row_min_df, new_mst_edges

def draw_net(adata, mst, reduce_dim_name):
    nodesunique = adata.obs['GroupId'].unique()
    item1Index = [item2 for item2 in nodesunique]
    cmap = plt.cm.Spectral
    bounds = range(min(item1Index), max(item1Index)+len(item1Index))
    norm = mcolors.BoundaryNorm(bounds, cmap.N)
    node_colors = [cmap(norm(value)) for value in adata.obs['GroupId'][list(mst.nodes)].values]
    pos = nx.spring_layout(mst)
    adata_copy = adata.copy()
    adata_copy = adata_copy[list(pos.keys()), :]
    pos_copy = {}
    i = 0
    for item in list(pos.keys()):
        pos_copy[item] = adata_copy.obsm[reduce_dim_name][i, :]
        i += 1
    plt.figure(figsize=(14, 10))
    nx.draw_networkx_nodes(mst, pos_copy, node_size=30, alpha=1, node_color=node_colors)
    nx.draw_networkx_edges(mst, pos_copy, edge_color='k', width=1, alpha=0.3, arrowsize=3)
    # nx.draw_networkx_labels(mst, pos_copy, font_size=7, font_color='k', font_family='Times New Roman')
    # plt.title('True Milestone Network')
    plt.axis('off')
    # plt.savefig(result_path+FolderName+'/mstfig.svg',format='svg')
    plt.show()

def read_GRN(svg_express_array, spot_num, hsvg_num, folder_way):
    '''通过读取文件获得GRNS相关信息'''
    GRNs_dim = np.empty([spot_num, hsvg_num])####存储GRN的度（基因对外调控，即基因的重要性）
    GRNs = np.empty([len(svg_express_array), len(svg_express_array[0]), len(svg_express_array[0])], dtype='float32')#初始化一个array来存储所有spot的GRN
    for i in range(len(svg_express_array)):
        file_way = folder_way + '/cell' + str(i+1) + '.txt'
        GRNs[i] = np.loadtxt(file_way)
        for j in range(len(GRNs[i])):
            GRNs[i][j][j] = 0
        #GRNs[np.where(np.abs(GRNs) < 0.01)] = 0######将GRNs的小于0.01的值设置为0
        GRNs_dim[i] = np.sum(np.abs(GRNs[i]), axis = 0)
    return GRNs, GRNs_dim

def GRN_in_out(GRNs, gene):
    # 获得GRN的in和out
    gene1 = np.empty((len(GRNs), 2))
    i = 0
    for item in GRNs:
        dim_in = np.sum(np.abs(item), axis = 1)
        dim_out = np.sum(np.abs(item), axis = 0)
        gene1[i, 0] = dim_in[gene]
        gene1[i, 1] = dim_out[gene]
        i+=1
    return gene1


def main(data_path, FolderName, result_path):
    # Load the RDS file
    rds_data = robjects.r['readRDS'](data_path + FolderName + '.rds')
    # 将R中的稀疏矩阵转换为Python中的稀疏矩阵
    dense_matrix = robjects.r['as'](rds_data.rx2('expression'), 'matrix')
    # 将稠密矩阵转换为NumPy数组
    dense_array = np.array(dense_matrix)
    # 获取稀疏矩阵的行名和列名
    # rownames = list(rds_data.rx2('cell_info')[0])
    # colnames = list(rds_data.rx2('feature_info')[0])
    rownames = list(rds_data.rx2('cell_info').rx2('cell_id'))
    colnames = list(rds_data.rx2('feature_info').rx2('feature_id'))
    adata = ad.AnnData(dense_array, obs = pd.DataFrame(index = rownames), var = pd.DataFrame(index = colnames))
    adata.obs['group_id'] = rds_data.rx2('grouping')
    adata.uns['milestone_network'] = rds_data.rx2('milestone_network')
    adata.uns['start_id'] = list(rds_data.rx2('prior_information').rx2('start_id'))
    adata.uns['start_milestones'] = list(rds_data.rx2('prior_information').rx2('start_milestones'))
    adata.obs['stage'] = np.array(rds_data.rx2('prior_information').rx2('timecourse_discrete'))
    # adata.obs['stage'] = (adata.obs['stage'].values-np.min(adata.obs['stage']))/(np.max(adata.obs['stage'])-np.min(adata.obs['stage']))
    sc.pp.normalize_total(adata, target_sum=1e4)
    sc.pp.log1p(adata)
    try:
        sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
        sc.pp.filter_genes_dispersion(adata, n_top_genes = 2000)
    except:
        print('Have not enouth genes')
    neighborsNumber = int(len(adata.obs) * 0.1)
    print(adata.uns['milestone_network'])
    pooled_data = pd.read_csv(result_path+FolderName+'/pooled_data.csv', index_col=0)
    adata.obsm['pooled_data'] = pooled_data.values
    GroupIdDict = {}
    i = 0
    Group_cell_numbers = {}
    for item in adata.obs['group_id'].unique():
        GroupIdDict[item] = i
        Group_cell_numbers[item] = 0
        i+=1
    GroupId = []
    for item in adata.obs['group_id'].values:
        GroupId.append(int(GroupIdDict[item]))
        Group_cell_numbers[item] += 1
    adata.obs['GroupId'] = GroupId
    Group_cell_numbers_df = pd.DataFrame(index=Group_cell_numbers.keys(), columns=['cell numbers'])
    for item in Group_cell_numbers.keys():
        Group_cell_numbers_df.loc[item, 'cell numbers'] = Group_cell_numbers[item]
    Group_cell_numbers_df.to_csv('D:/A_study/A_study/cell_differentiation2/paper/paper/fig/result/'+FolderName + '/Group_cell_numbers.csv')


    sc.tl.pca(adata, svd_solver="arpack")
    # pca_data = adata.obsm['X_pca']
    sc.pp.neighbors(adata, n_neighbors=neighborsNumber)
    sc.tl.umap(adata)

    distance_all = np.loadtxt(result_path+FolderName+'/distance_all.txt')


    _, eigenvectors_distance_all = np.linalg.eig(distance_all)
    eigenvectors_distance_all = np.real(eigenvectors_distance_all)
    adata.obsm['X_eigenvectors'] = eigenvectors_distance_all[:, :2]
    # adata.obsm['X_eigenvectors'][:,0] = adata.obsm['X_eigenvectors'][:,0]/np.max(adata.obsm['X_eigenvectors'][:,0])
    # adata.obsm['X_eigenvectors'][:,1] = adata.obsm['X_eigenvectors'][:,1]/np.max(adata.obsm['X_eigenvectors'][:,1])
    # adata.obsm['X_umap'][:,0] = adata.obsm['X_umap'][:,0]/np.max(adata.obsm['X_umap'][:,0])
    # adata.obsm['X_umap'][:,1] = adata.obsm['X_umap'][:,1]/np.max(adata.obsm['X_umap'][:,1])
    adata.obsm['X_dif'] = adata.obsm['X_eigenvectors']/np.max(adata.obsm['X_eigenvectors']) + adata.obsm['X_umap'] / np.max(adata.obsm['X_umap'] )


    # adata.obsm['distance_all'] = distance_all
    RootGroup1 = []
    root_index = [np.where(adata.obs.index==item)[0][0] for item in list(rds_data.rx2('prior_information').rx2('start_id'))]
    for item in adata.obs['GroupId'].values[root_index]:
        RootGroup1.append(item)


    pseudo_time = np.loadtxt(result_path+FolderName+'/pseudo_time.txt')
    adata.obs['pseudo time'] = pseudo_time

    GroupIndex = []
    for item in adata.obs['GroupId'].unique():
        GroupIndex.append(np.where(adata.obs['GroupId'] == item)[0])
    # G_amend_edges = np.loadtxt(result_path+FolderName+'\G_edges.txt', dtype=str)
    # G_amend_edges = np.loadtxt(result_path+FolderName+'\G_edges.txt', dtype=str, delimiter=',')
    # G_amend_edges_list = []
    # for i in range(int(len(G_amend_edges)/3)):
    #     G_amend_edges_list.append((G_amend_edges[i*3+0], G_amend_edges[i*3+1], {'weight':eval(G_amend_edges[i*3+2])}))
    # G = nx.Graph()
    # G.add_edges_from(G_amend_edges_list)
    # G_mst = nx.minimum_spanning_tree(G, algorithm='kruskal')
    # mst_Group_edges = np.loadtxt(result_path+FolderName+'\mst_Group.txt', dtype=str, comments=None, delimiter=',')
    mst_Group_edges = np.loadtxt(result_path+FolderName+'\mst_Group.txt', dtype=str, comments=None)
    # print(mst_Group_edges)
    mst_Group_edges_list = []
    print(mst_Group_edges)
    for i in range(int(len(mst_Group_edges))):
        mst_Group_edges_list.append((mst_Group_edges[i, 0], mst_Group_edges[i, 1]))
    Group_mst = nx.Graph()
    Group_mst.add_edges_from(mst_Group_edges_list)
    start_Group = adata.uns['start_milestones']
    pseudo_time_Group, new_mst_edges = shortest_path_to_pseudo_time_group(Group_mst, start_Group)
    # print(pseudo_time_Group)
    # print(adata.obs['pseudo time'])
    for item in adata.obs.index:
        adata.obs.loc[item, 'pseudo time'] = adata.obs.loc[item, 'pseudo time'] + pseudo_time_Group.loc[adata.obs.loc[item, 'group_id'], 'pseudo time']
    # resultplt(adata.obsm['X_dif'], adata.obs['pseudo time'].values)
    mean_position_umap = {}
    for item in adata.obs['group_id'].unique():
        indices = adata.obs.index[adata.obs['group_id'] == item].tolist()
        # adata_copy = adata[indices, ].obsm['X_dif']
        position = adata[indices, ].obsm['X_umap']
        mean_position_umap[item] = np.mean(position, axis=0)
    mean_position_dif = {}
    for item in adata.obs['group_id'].unique():
        indices = adata.obs.index[adata.obs['group_id'] == item].tolist()
        # adata_copy = adata[indices, ].obsm['X_dif']
        position = adata[indices, ].obsm['X_dif']
        mean_position_dif[item] = np.mean(position, axis=0)
    
    # --------------------------轨迹图--------------------------
    fig1, ax = plt.subplots()
    colors = ['#1E469B','#2681B6', '#35B9C5','#96D2B0','#F9F8CA']
    cmap = LinearSegmentedColormap.from_list("custom_gradient", colors)
    scatter=ax.scatter(adata.obsm['X_dif'][:,0], adata.obsm['X_dif'][:,1],c=adata.obs['pseudo time'].values, cmap=cmap)
    # plt.colorbar(scatter, ax=ax)
    for item in new_mst_edges:
        start_x, start_y = mean_position_dif[item[0]][0], mean_position_dif[item[0]][1]
        end_x, end_y = mean_position_dif[item[1]][0], mean_position_dif[item[1]][1]
        dx = end_x - start_x
        dy = end_y - start_y
        # ax.arrow(start_x, start_y, dx, dy, head_width=0.1, head_length=0.1, fc='k', ec='k')
        font_properties = {
            'fontsize': 10,
            'fontweight': 'bold',
            'fontstyle': 'italic',
            'fontname': 'Times New Roman',
            'color': 'black'
        }
        ax.text(start_x, start_y, item[0], fontdict=font_properties, verticalalignment='bottom', horizontalalignment='right')
        ax.text(end_x, end_y, item[1], fontdict=font_properties, verticalalignment='bottom', horizontalalignment='right')
        plt.plot([start_x,end_x], [start_y,end_y], c = 'k')
    ax.axis('off')
    save_dir = 'D:/A_study/A_study/cell_differentiation2/paper/paper/fig/result/'+FolderName
    os.makedirs(save_dir, exist_ok=True)
    # 创建图像文件的完整路径
    save_path = os.path.join(save_dir, 'Differentiation_trajectory.svg')
    plt.savefig(save_path, format='svg')
    plt.show()

    # fig2, ax = plt.subplots()
    # colors = ['#1E469B','#2681B6', '#35B9C5','#96D2B0','#F9F8CA']
    # cmap = LinearSegmentedColormap.from_list("custom_gradient", colors)
    # print(cmap)
    # scatter=ax.scatter(adata.obsm['X_umap'][:,0], adata.obsm['X_umap'][:,1],c=adata.obs['pseudo time'].values, cmap=cmap)
    # # plt.colorbar(scatter, ax=ax)
    # for item in new_mst_edges:
    #     start_x, start_y = mean_position_umap[item[0]][0], mean_position_umap[item[0]][1]
    #     end_x, end_y = mean_position_umap[item[1]][0], mean_position_umap[item[1]][1]
    #     plt.plot([start_x,end_x], [start_y,end_y], c = 'k')
    # ax.axis('off')
    # save_dir = 'D:/A_study/A_study/cell_differentiation2/paper/paper/fig/result/'+FolderName
    # os.makedirs(save_dir, exist_ok=True)
    # # 创建图像文件的完整路径
    # save_path = os.path.join(save_dir, 'Differentiation_trajectory2.svg')
    # # plt.savefig(save_path, format='svg')
    # plt.show()

    # fig2, ax = plt.subplots()
    # colors = ['#1E469B','#2681B6', '#35B9C5','#96D2B0','#F9F8CA']
    # cmap = LinearSegmentedColormap.from_list("custom_gradient", colors)
    # print(cmap)
    # scatter=ax.scatter(adata.obsm['X_dif'][:,0], adata.obsm['X_dif'][:,1],c=adata.obs['pseudo time'].values, cmap=cmap)
    # # plt.colorbar(scatter, ax=ax)
    # for item in new_mst_edges:
    #     start_x, start_y = mean_position_dif[item[0]][0], mean_position_dif[item[0]][1]
    #     end_x, end_y = mean_position_dif[item[1]][0], mean_position_dif[item[1]][1]
    #     plt.plot([start_x,end_x], [start_y,end_y], c = 'k')
    # ax.axis('off')
    # save_dir = 'D:/A_study/A_study/cell_differentiation2/paper/paper/fig/result/'+FolderName
    # os.makedirs(save_dir, exist_ok=True)
    # # 创建图像文件的完整路径
    # save_path = os.path.join(save_dir, 'Differentiation_trajectory2.svg')
    # # plt.savefig(save_path, format='svg')
    # plt.show()
    # --------------------------聚类图--------------------------
    # GroupIdDict = {}
    # i = 0
    # for item in adata.obs['group_id'].unique():
    #     GroupIdDict[item] = i
    #     i+=1
    # labels = list(GroupIdDict.keys())

    # fig3, ax = plt.subplots()
    # original_colors = ['#55b7e6', '#56ba77', '#ed6e69', '#9faa3f', '#bd79b6']
    # cmap = LinearSegmentedColormap.from_list("custom_cmap", colors)

    # # 提取离散颜色
    # num_colors = len(labels)
    # discrete_colors = [to_hex(cmap(i / num_colors)) for i in range(num_colors)]

    # # 创建标签到颜色的映射
    # label_color_mapping = {label: color for label, color in zip(labels, discrete_colors)}
    # x, y = adata.obsm['X_dif'][:,0], adata.obsm['X_dif'][:,1]
    # print(labels)
    # # 创建散点图并按标签分组
    # for label in labels:
    #     # 筛选出当前标签对应的数据
    #     x_label = [x[i] for i, l in enumerate(adata.obs['group_id']) if l == label]
    #     y_label = [y[i] for i, l in enumerate(adata.obs['group_id']) if l == label]
    #     # 绘制散点图
    #     color = label_color_mapping[label]
    #     plt.scatter(x_label, y_label, c=color, label=label)

    # # 添加图例
    # plt.legend(frameon=False, facecolor='none')

    # for item in new_mst_edges:
    #     start_x, start_y = mean_position[item[0]][0], mean_position[item[0]][1]
    #     end_x, end_y = mean_position[item[1]][0], mean_position[item[1]][1]
    #     dx = end_x - start_x
    #     dy = end_y - start_y
    #     font_properties = {
    #         'fontsize': 10,
    #         'fontweight': 'bold',
    #         'fontstyle': 'italic',
    #         'fontname': 'Times New Roman',
    #         'color': 'black'
    #     }
    #     ax.text(start_x, start_y, item[0], fontdict=font_properties, verticalalignment='bottom', horizontalalignment='right')
    #     ax.text(end_x, end_y, item[1], fontdict=font_properties, verticalalignment='bottom', horizontalalignment='right')
    # ax.axis('off')
    # # 保存图像为SVG格式
    # save_dir = 'D:/A_study/A_study/cell_differentiation2/paper/paper/fig/result/'+FolderName
    # os.makedirs(save_dir, exist_ok=True)
    # save_path = os.path.join(save_dir, 'cluster.svg')
    # plt.savefig(save_path, format='svg')
    # plt.show()

    # fig4, ax = plt.subplots()
    # original_colors = ['#55b7e6', '#56ba77', '#ed6e69', '#9faa3f', '#bd79b6']
    # cmap = LinearSegmentedColormap.from_list("custom_cmap", colors)

    # # 提取离散颜色
    # num_colors = len(labels)
    # discrete_colors = [to_hex(cmap(i / num_colors)) for i in range(num_colors)]

    # # 创建标签到颜色的映射
    # label_color_mapping = {label: color for label, color in zip(labels, discrete_colors)}
    # x, y = adata.obsm['X_dif'][:,0], adata.obsm['X_dif'][:,1]
    # # 创建散点图并按标签分组
    # for label in labels:
    #     # 筛选出当前标签对应的数据
    #     x_label = [x[i] for i, l in enumerate(adata.obs['group_id']) if l == label]
    #     y_label = [y[i] for i, l in enumerate(adata.obs['group_id']) if l == label]
    #     # 绘制散点图
    #     color = label_color_mapping[label]
    #     plt.scatter(x_label, y_label, c=color, label=label)

    # # 添加图例
    # # plt.legend(frameon=False, facecolor='none')
    # ax.axis('off')
    # # 保存图像为SVG格式
    # save_dir = 'D:/A_study/A_study/cell_differentiation2/paper/paper/fig/result/'+FolderName
    # os.makedirs(save_dir, exist_ok=True)
    # save_path = os.path.join(save_dir, 'cluster2.svg')
    # plt.savefig(save_path, format='svg')
    # plt.show()

    # fig5, ax = plt.subplots()
    # original_colors = ['#55b7e6', '#56ba77', '#ed6e69', '#9faa3f', '#bd79b6']
    # cmap = LinearSegmentedColormap.from_list("custom_cmap", colors)

    # # 提取离散颜色
    # num_colors = len(labels)
    # discrete_colors = [to_hex(cmap(i / num_colors)) for i in range(num_colors)]

    # # 创建标签到颜色的映射
    # label_color_mapping = {label: color for label, color in zip(labels, discrete_colors)}
    # x, y = adata.obsm['X_dif'][:,0], adata.obsm['X_dif'][:,1]
    # # 创建散点图并按标签分组
    # for label in labels:
    #     # 筛选出当前标签对应的数据
    #     x_label = [x[i] for i, l in enumerate(adata.obs['group_id']) if l == label]
    #     y_label = [y[i] for i, l in enumerate(adata.obs['group_id']) if l == label]
    #     # 绘制散点图
    #     color = label_color_mapping[label]
    #     plt.scatter(0, 0, c=color, label=label)

    # # 添加图例
    # plt.legend(frameon=False, facecolor='none')
    # ax.axis('off')
    # # 保存图像为SVG格式
    save_dir = 'D:/A_study/A_study/cell_differentiation2/paper/paper/fig/result/'+FolderName
    # os.makedirs(save_dir, exist_ok=True)
    # save_path = os.path.join(save_dir, 'legend.svg')
    # plt.savefig(save_path, format='svg')
    # plt.show()
    # # --------------------------基因表达热图--------------------------
    # plt.figure()
    # pearson_corr_list = []
    # gene_names = list(adata.var.index)
    # for i in range(len(adata.X[0,:])):
    #     pearson_corr_list.append(PearsonCorr_func(adata.obs['pseudo time'].values, adata.X[:,i]))
    # gene_PT_pearson_corr = pd.DataFrame(pearson_corr_list, index=gene_names, columns=['pearson corr'])
    # gene_PT_pearson_corr_sorted = gene_PT_pearson_corr.sort_values(by='pearson corr', ascending=True)
    # import fitter_p_value
    # p_value = fitter_p_value.calculate_threshold_gene_site(gene_PT_pearson_corr_sorted.values)
    # gene_PT_pearson_corr_sorted['p_value'] = p_value
    # save_path = os.path.join(save_dir, 'gene_PT_pearson_corr.csv')
    # gene_PT_pearson_corr_sorted.to_csv(save_path)
    # save_path = os.path.join(save_dir, 'gene_PT_pearson_corr_positive.csv')
    # gene_PT_pearson_corr_sorted[gene_PT_pearson_corr_sorted['p_value'] < 0.05].to_csv(save_path)
    # save_path = os.path.join(save_dir, 'gene_PT_pearson_corr_negative.csv')
    # gene_PT_pearson_corr_sorted[gene_PT_pearson_corr_sorted['p_value'] > 0.95].to_csv(save_path)
    # # adata_copy = adata.copy()
    # PT_sorted = adata.obs.sort_values(by='pseudo time', ascending=True).index
    # adata_copy = adata[PT_sorted,:]
    # imshow_array1 = adata_copy[:, gene_PT_pearson_corr_sorted.index[:20]].X
    # imshow_array2 = adata_copy[:, gene_PT_pearson_corr_sorted.index[-20:]].X
    # plt_Genes = list(gene_PT_pearson_corr_sorted.index[:20]) + list(gene_PT_pearson_corr_sorted.index[-20:])
    # save_dir = 'D:/A_study/A_study/cell_differentiation2/paper/paper/fig/result/'+FolderName
    # os.makedirs(save_dir, exist_ok=True)
    # save_path = os.path.join(save_dir, 'plt_Genes.txt')
    # np.savetxt(save_path, plt_Genes, fmt='%s')
    # imshow_array = np.hstack((imshow_array1, imshow_array2))
    # if len(adata) < 100:
    #     imshow_array = imshow_array
    # else:
    #     imshow_array = imshow_array[::int(len(adata)/100)]
    # plt.figure()
    # colors = ['#276C9E','#A3C9D5', '#E3EA96','#FCDC89','#E26844', '#8A233F']
    # cmap = LinearSegmentedColormap.from_list("custom_gradient", colors)
    # plt.imshow(imshow_array.T, cmap=cmap)
    # plt.axis('off')
    # save_path = os.path.join(save_dir, 'GeneExpr.svg')
    # plt.savefig(save_path, format='svg')
    # plt.show()
    
    # PT_sorted = adata.obs.sort_values(by='pseudo time', ascending=True).index
    # adata_copy = adata[PT_sorted,:]
    # pearson_corr_list = []
    # gene_names = list(pooled_data.columns)
    # for i in range(len(adata_copy.obsm['pooled_data'][0, :])):
    #     pearson_corr_list.append(PearsonCorr_func(adata_copy.obs['pseudo time'].values, adata_copy.obsm['pooled_data'][:,i]))
    # gene_PT_pearson_corr = pd.DataFrame(pearson_corr_list, index=gene_names, columns=['pearson corr'])
    # gene_PT_pearson_corr_sorted = gene_PT_pearson_corr.sort_values(by='pearson corr', ascending=True)
    # pooled_data_copy = pd.DataFrame(adata_copy.obsm['pooled_data'].T, index=gene_names, columns=adata_copy.obs.index)
    # pooled_data_copy_reindexed = pooled_data_copy.reindex(gene_PT_pearson_corr_sorted.index)
    # # pooled_data_copy_copy = pooled_data_copy[[item for item in gene_PT_pearson_corr_sorted.index]]
    # plt.figure()
    # imshow_array = pooled_data_copy_reindexed.values.T
    # if len(adata) < 100:
    #     imshow_array = imshow_array
    # else:
    #     imshow_array = imshow_array[::int(len(adata)/100)]
    # plt.figure()
    # colors = ['#276C9E','#A3C9D5', '#E3EA96','#FCDC89','#E26844', '#8A233F']
    # cmap = LinearSegmentedColormap.from_list("custom_gradient", colors)
    # plt.imshow(imshow_array.T, cmap=cmap)
    # plt.axis('off')
    # save_dir = 'D:/A_study/A_study/cell_differentiation2/paper/paper/fig/result/'+FolderName
    # save_path = os.path.join(save_dir, 'PooledGeneExpr.svg')
    # plt.savefig(save_path, format='svg')
    # plt.show()

    # --------------------------基因调控图--------------------------
    # hvg_express_array = pooled_data.values
    # GRNs, _ = read_GRN(hvg_express_array, len(hvg_express_array), len(hvg_express_array[0]), result_path + FolderName)
    # for i in range(len(hvg_express_array[0])):
    #     gene1 = GRN_in_out(GRNs, i)
    #     plt.figure()
    #     colors = ['#1E469B','#2681B6', '#35B9C5','#96D2B0','#F9F8CA']
    #     cmap = LinearSegmentedColormap.from_list("custom_gradient", colors)
    #     ax = plt.gca()
    #     ax.spines['top'].set_visible(False)
    #     ax.spines['right'].set_visible(False)
    #     plt.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False, labelbottom=False, labelleft=False)
    #     # 添加网格线（可选）
    #     plt.grid(True, which='both', color='gray', linestyle='-', linewidth=2, alpha = 0.2)
    #     plt.annotate('', xy=(1.05, 0), xytext=(-0.05, 0),
    #                 arrowprops=dict(arrowstyle="->", color='black', lw=2, mutation_scale=30),
    #                 xycoords=('axes fraction', 'axes fraction'), textcoords=('axes fraction', 'axes fraction'))
    #     plt.annotate('', xy=(0, 1.05), xytext=(0, -0.05),
    #                 arrowprops=dict(arrowstyle="->", color='black', lw=2, mutation_scale=30),
    #                 xycoords=('axes fraction', 'axes fraction'), textcoords=('axes fraction', 'axes fraction'))
    #     plt.scatter(gene1[:, 0], gene1[:, 1], c = adata.obs['pseudo time'], cmap = cmap)
    #     plt.savefig('D:/A_study/A_study/cell_differentiation2/paper/paper/fig/result/'+FolderName+'/dimIO_pseudo time'+str(i+1)+'.svg', format='svg')
    #     plt.show()

    #     plt.figure()
    #     colors = ['#276C9E','#A3C9D5', '#E3EA96','#FCDC89','#E26844', '#8A233F']
    #     cmap = LinearSegmentedColormap.from_list("custom_gradient", colors)
    #     ax = plt.gca()
    #     ax.spines['top'].set_visible(False)
    #     ax.spines['right'].set_visible(False)
    #     plt.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False, labelbottom=False, labelleft=False)
    #     # 添加网格线（可选）
    #     plt.grid(True, which='both', color='gray', linestyle='-', linewidth=2, alpha = 0.2)
    #     plt.annotate('', xy=(1.05, 0), xytext=(-0.05, 0),
    #                 arrowprops=dict(arrowstyle="->", color='black', lw=2, mutation_scale=30),
    #                 xycoords=('axes fraction', 'axes fraction'), textcoords=('axes fraction', 'axes fraction'))
    #     plt.annotate('', xy=(0, 1.05), xytext=(0, -0.05),
    #                 arrowprops=dict(arrowstyle="->", color='black', lw=2, mutation_scale=30),
    #                 xycoords=('axes fraction', 'axes fraction'), textcoords=('axes fraction', 'axes fraction'))
    #     plt.scatter(gene1[:, 0], gene1[:, 1], c = pooled_data.iloc[:, i].values, cmap=cmap)
    #     # plt.title('gene'+str(i+1))
    #     # plt.colorbar()
    #     plt.savefig('D:/A_study/A_study/cell_differentiation2/paper/paper/fig/result/'+FolderName+'/dimIO_GeneExpr'+str(i+1)+'.svg', format='svg')
    #     plt.show()
    # 创建一个图形和一个轴，只用于显示颜色条
    # gradient = [cmap(i) for i in range(256)]
    # # Plot the gradient
    # fig, ax = plt.subplots(figsize=(6, 1))
    # ax.imshow([gradient], aspect='auto')
    # ax.set_axis_off()
    # plt.savefig('D:/A_study/A_study/cell_differentiation2/paper/paper/fig/result/colorbar_GeneExpr.svg', format='svg')
    # plt.show()
    
```
